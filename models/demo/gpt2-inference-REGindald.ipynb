```python
# Load the trained model from the disk
from transformers import AutoModelForCausalLM

model_dir = './drive/MyDrive/REGindald-trained/reginald-distilgpt2-10/' if IN_COLAB else './reginald-distilgpt2-10/'
saved_model = AutoModelForCausalLM.from_pretrained(model_dir, local_files_only=True)
```

```python
from transformers import AutoTokenizer

# We also need to get the tokenizer again!
checkpoint = 'distilgpt2'
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

```

```python
import torch
from transformers import pipeline

# Using text-generation pipeline
# Hugging Face has pipelines for all standardized tasks

device = "cuda:0" if torch.cuda.is_available() else "cpu"

pipe = pipeline(
    "text-generation", 
    model=saved_model, 
    device=device, 
    tokenizer=tokenizer
)
```

```python
def generate_and_print(inputs, n, output_length):
  outputs = pipe(inputs, num_return_sequences=n, max_new_tokens=output_length)
  for x in outputs[0]:
    print(x['generated_text'])
  
generate_and_print(["In the Research Engineering Group in the Alan Turing Institute, Evelina"], 10,50)  
```
